```python
climate_prod_df = pd.merge(potato_prod, climate, on='Year')
climate_prod_df
```

```python
from sklearn.preprocessing import StandardScaler
# Standardize all the columns except the Year
scaler = StandardScaler()
cols_to_scale = [col for col in climate_prod_df.columns if col not in ['Year','Production']]
climate_prod_df[cols_to_scale] = scaler.fit_transform(climate_prod_df[cols_to_scale])
climate_prod_df
```

```python
corr_prod = climate_prod_df.corrwith(climate_prod_df[('Production')])

fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr_prod.to_frame(), cmap='coolwarm', annot=True, ax=ax)

# Set the title of the plot
ax.set_title('Correlation between Production and other variables')
```

```python
# Filter columns with correlation >= 0.5
high_corr_cols = corr_prod[abs(corr_prod) >= 0.5].index.tolist()

# Add Year column
final_cols = ['Year'] + high_corr_cols


# Drop columns with low correlation
climate_prod_df = climate_prod_df.loc[:, final_cols]
climate_prod_df
```

```python
corr_prod = climate_prod_df.corrwith(climate_prod_df[('Production')])

fig, ax = plt.subplots(figsize=(10,7))
sns.heatmap(corr_prod[1:].to_frame(), cmap='coolwarm', annot=True, ax=ax)

# Set the title of the plot
ax.set_title('Correlation between Production and other variables')
```
