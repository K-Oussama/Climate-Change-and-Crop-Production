```python
import pandas as pd
```

```python
# Load the dataset
climate = pd.read_csv("../../Data/climate-change_mar.csv")
climate.head()
```

```python
# drop the first row
climate = climate.drop(index=0)
# reset the index
climate = climate.reset_index(drop=True)
climate
```

```python
indicators = climate[['Indicator Name','Indicator Code']].copy()
indicators = indicators.astype(str)
indicators.drop_duplicates(inplace=True)
indicators = indicators.reset_index(drop=True)
indicators.to_csv('../../Data/climate-indicators_mapping.csv', index=False)
indicators
```

```python
climate = climate.drop(columns=["Country Name","Country ISO3","Indicator Name"])
climate.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Pivot the dataframe
df_pivot = climate.pivot(index='Year', columns='Indicator Code', values='Value')

# Rename the columns to remove whitespace
df_pivot.columns = df_pivot.columns.str.replace(' ', '_')

# Reset the index
climate = df_pivot.reset_index()

climate.head()
```

```python
#del climate["Indicator Code"]
#climate = climate.drop(columns='Indicator_Code')
climate.columns
```

```python
climate.info()
```

```python
# Change the data type of all columns except "Year" to float
climate[climate.columns.difference(['Year'])] = climate[climate.columns.difference(['Year'])].astype(float)

# Change the data type of "Year" column to int
climate['Year'] = climate['Year'].astype(int)

climate.info()
```

```python
climate.Year.unique()
```

```python
sum(climate.Year.duplicated())
```

```python
climate = climate[(climate['Year'] >= 2011) & (climate['Year'] <= 2020)]
print(climate.Year.unique())
climate
```

```python
duplicates = climate.duplicated()
print("Number of duplicated rows:", duplicates.sum())
```

```python
low_count_columns = climate.columns[climate.count() < 7]
low_count_columns
```

```python
low_count_indicators = indicators.loc[indicators['Indicator Code'].isin(low_count_columns)]
print("Number of the indicators that need to be deleted:",low_count_indicators['Indicator Name'].drop_duplicates().count())
low_count_indicators.drop_duplicates()
```

```python
climate = climate.drop(columns=low_count_columns)
```

```python
# Replace NaN values with column mean
climate.fillna(climate.mean(), inplace=True)
```

```python
same_val_cols = [col for col in climate.columns if climate[col].nunique() == 1]
print(same_val_cols)
climate = climate.drop(same_val_cols, axis=1)
```
