```python
import pandas as pd
```

```python
prodv = pd.read_excel("../../Data/production-vegetale.xlsx")
prodv
```

```python
prodv.columns
```

```python
# Remove leading white spaces from the column name
prodv.columns = prodv.columns.str.strip()
```

```python
# Split the Occurrence column into Year and remove the first 4 characters
prodv['Year'] = prodv['Occurrence'].str[-4:].astype(int)
prodv.drop('Occurrence', axis=1, inplace=True)
```

```python
# Pivot the dataframe
df_pivot = pd.pivot_table(prodv, index=['Year', 'Filière', 'Produit'], columns='Indicateur', values='Valeur')
df_pivot
```

```python
# Reset the index and rename columns
df_pivot = df_pivot.reset_index()
df_pivot.columns.name = None
prodv = df_pivot.rename(columns={'Production-T (Tonnes)': 'Production', 'Superficie (Ha)': 'Superficie'})
prodv
```

```python
# Reorder the columns
prodv = prodv[['Year', 'Filière', 'Produit', 'Production', 'Superficie']]
prodv.columns
```

```python
print("Number of years : ",len(prodv.Year.unique()))
prodv.Year.unique()
```
